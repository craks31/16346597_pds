```python
import pandas as pd
import matplotlib.pyplot as pl
```

```python
# Specify the file path to your CSV file
cleaned_file_path = '/content/drive/Othercomputers/My PC/PDS A1/StudentPerformance Project/clean_data_PHASE2/modified_studentPerformance_data.csv'

# Use pandas to read the CSV file into a DataFrame
df = pd.read_csv(cleaned_file_path)

output_txt_file = '/content/drive/Othercomputers/My PC/PDS A1/StudentPerformance Project/results_PHASE3/results.txt'
resultFile = open(output_txt_file, 'w')

output_folder = '/content/drive/Othercomputers/My PC/PDS A1/StudentPerformance Project/results_PHASE3/'



```

```python

# 1. Box Plot of Reading Scores by Gender
pl.figure(figsize=(8, 6))
df.boxplot(column='reading score', by='gender', grid=False)
pl.title('Box Plot of Reading Scores by Gender')
pl.xlabel('Gender')
pl.ylabel('Reading Score')
pl.suptitle('')  # Remove default title


# Define the file name (e.g., 'reading_scores_by_gender.png')
file_name = 'box_plot_of_reading_scores_by_gender.png'

# Combine the folder path and file name
output_path = output_folder + file_name

# Save the plot as an image file in the specified folder
pl.savefig(output_path)
pl.show()
resultFile.write("\nAnalysis Simplification for Viusalisation 1\n")
resultFile.write("\nBox Plot of Reading Scores by Gender\n")
resultFile.write(str('The box plot provides a visual comparison of the distribution of reading scores between male and female students. This comparison allows you to quickly assess whether there are differences in the central tendency and spread of scores between the two gender groups.'+
'Identification of Key Statistics: The box plot displays important summary statistics, such as the median (the line inside the box), the interquartile range (the height of the box), and potential outliers (individual data points beyond the "whiskers").'))
```

```python
#2

pl.figure(figsize=(12, 4))

# Math Scores
pl.subplot(131)
pl.hist(df[df['gender'] == 'male']['math score'], alpha=0.5, label='Male', color='blue', bins=15)
pl.hist(df[df['gender'] == 'female']['math score'], alpha=0.5, label='Female', color='pink', bins=15)
pl.title('Math Scores by Gender')
pl.xlabel('Math Score')
pl.ylabel('Frequency')
pl.legend()

# Reading Scores
pl.subplot(132)
pl.hist(df[df['gender'] == 'male']['reading score'], alpha=0.5, label='Male', color='blue', bins=15)
pl.hist(df[df['gender'] == 'female']['reading score'], alpha=0.5, label='Female', color='pink', bins=15)
pl.title('Reading Scores by Gender')
pl.xlabel('Reading Score')
pl.ylabel('Frequency')
pl.legend()

# Writing Scores
pl.subplot(133)
pl.hist(df[df['gender'] == 'male']['writing score'], alpha=0.5, label='Male', color='blue', bins=15)
pl.hist(df[df['gender'] == 'female']['writing score'], alpha=0.5, label='Female', color='pink', bins=15)
pl.title('Writing Scores by Gender')
pl.xlabel('Writing Score')
pl.ylabel('Frequency')
pl.legend()

pl.tight_layout()


# Define the file name (e.g., 'reading_scores_by_gender.png')
file_name = 'Histograms of Exam Scores by Gender.png'

# Combine the folder path and file name
output_path = output_folder + file_name

# Save the plot as an image file in the specified folder
pl.savefig(output_path)
pl.show()
resultFile.write("\nAnalysis Simplification for Viusalisation 2\n")
resultFile.write("\nHistograms of Exam Scores by Gender\n")

resultFile.write(str('Analysis Simplification: By visualizing the distribution of scores separately for male and female students, you can quickly compare the performance of the two groups. This separation makes it easier to identify differences or similarities in exam scores between genders. It simplifies the analysis by breaking it down into two distinct groups'+
'Understanding Score Distributions: Histograms provide a clear overview of the score distributions. You can see the central tendency (e.g., where the majority of scores lie) and the spread (e.g., the range of scores) for each subject. This helps in understanding the typical performance of male and female students in each subject.'+
'Detecting Gender-Based Patterns: If there are notable patterns, such as differences in the shape of the distributions or shifts in the means, they become apparent in the histograms. For example, you might observe that female students tend to have higher reading scores than male students.'
))
```

```python
#3
import seaborn as sns
import matplotlib.pyplot as plt
correlation_matrix = df[['math score', 'reading score', 'writing score']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Exam Scores')


# Define the file name (e.g., 'reading_scores_by_gender.png')
file_name = 'Correlation Heatmap of Exam Scores.png'

# Combine the folder path and file name
output_path = output_folder + file_name

# Save the plot as an image file in the specified folder
plt.savefig(output_path)
plt.show()
resultFile.write("\nAnalysis Simplification for Viusalisation 3\n")
resultFile.write("\nCorrelation Heatmap of Exam Scores\n")
resultFile.write("The correlation heatmap provides a clear visual representation of the strength and direction of correlations between variables. "+
                 "It helps you prioritize variables that are strongly correlated. For example, if you're interested in predicting student performance, you can quickly identify which subjects (math, reading, or writing) have the strongest correlations with each other. ")
```

```python
#4


# Create subplots for math, reading, and writing scores
fig, axes = plt.subplots(1, 3, figsize=(30, 9), sharey=True)

# Plot math scores by parental education level
sns.boxplot(data=df, x='parental level of education', y='math score', palette='Set3', ax=axes[0])
axes[0].set_title('Math Scores')
axes[0].set_xlabel('Parental Level of Education')
axes[0].set_ylabel('Score')

# Plot reading scores by parental education level
sns.boxplot(data=df, x='parental level of education', y='reading score', palette='Set3', ax=axes[1])
axes[1].set_title('Reading Scores')
axes[1].set_xlabel('Parental Level of Education')
axes[1].set_ylabel('Score')

# Plot writing scores by parental education level
sns.boxplot(data=df, x='parental level of education', y='writing score', palette='Set3', ax=axes[2])
axes[2].set_title('Writing Scores')
axes[2].set_xlabel('Parental Level of Education')
axes[2].set_ylabel('Score')

# Adjust layout and show the plot
plt.tight_layout()


# Define the file name (e.g., 'reading_scores_by_gender.png')
file_name = 'Grouped box plots for analyzing how parental Education affects student performance.png'

# Combine the folder path and file name
output_path = output_folder + file_name

# Save the plot as an image file in the specified folder
plt.savefig(output_path)
plt.show()
resultFile.write("\nAnalysis Simplification for Viusalisation 4\n")
resultFile.write("\nGrouped box plots for analyzing how parental Education affects student performance\n")
resultFile.write("Each box plot provides insights into how different levels of parental education are associated with the distribution of scores in each subject. You can easily compare the medians, quartiles, and potential outliers for each education level.")
```

```python


# Set the figure size
plt.figure(figsize=(12, 6))

# Create a grouped box plot for math scores by race/ethnicity
sns.boxplot(data=df, x='race/ethnicity', y='math score', palette='Set3', width=0.4, showfliers=False)
plt.title('Math Scores by Race/Ethnicity')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Math Score')

# Show the plot
plt.show()

# Set the figure size
plt.figure(figsize=(12, 6))

# Create a grouped box plot for reading scores by race/ethnicity
sns.boxplot(data=df, x='race/ethnicity', y='reading score', palette='Set2', width=0.4, showfliers=False)
plt.title('Reading Scores by Race/Ethnicity')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Reading Score')

# Show the plot
plt.show()

# Set the figure size
plt.figure(figsize=(12, 6))

# Create a grouped box plot for writing scores by race/ethnicity
sns.boxplot(data=df, x='race/ethnicity', y='writing score', palette='Set1', width=0.4, showfliers=False)
plt.title('Writing Scores by Race/Ethnicity')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Writing Score')

# Show the plot


# Define the file name (e.g., 'reading_scores_by_gender.png')
file_name = 'Grouped box plots for analyzing how raceethnicity affects student performance.png'

# Combine the folder path and file name
output_path = output_folder + file_name

# Save the plot as an image file in the specified folder
plt.savefig(output_path)
plt.show()
resultFile.write("\nAnalysis Simplification for Viusalisation 5\n")
resultFile.write("\nGrouped box plots for analyzing how race/ethnicity affects student performance\n")
resultFile.write("Each box plot provides insights into how different groups of race/ethnicity are associated with the distribution of scores in each subject. You can easily compare the medians, quartiles, and potential outliers for each education level.")
resultFile.close()
```
